## 1. 퍼셉트론 (Perceptron)
---

인공 신경망 모형의 하나인 퍼셉트론(Perceptron)은 1957년에 프랑크 로젠블라트(Frank Rosenblatt)가 처음으로 제안한 알고리즘 입니다. 실제 뇌를 구성하는 신경 세포 뉴런의 동작을 본따 만들었습니다. 아래 그림과 같이 뉴런은 가지돌기에서 신호를 받아들이고, 이 신호가 일정치 이상의 크기를 가지면 축삭돌기를 통해서 신호를 전달합니다.



<p align='center'>
<img src=https://wikidocs.net/images/page/24958/%EB%89%B4%EB%9F%B0.PNG  width=450>
</p>



<h3>
<center>

$if \sum_i^{n} w_{i}x_{i} + b\ ≥ \theta → y=1$

$ if \sum_i^{n} w_{i}x_{i} +b\ < \theta → y=0$
</center>
</h3>

.


`x` 는 입력값을 의미하며, `w`는 **가중치**(Weight), `b`는 **편향** (bias), `y`는 출력값입니다. 

각각의 입력값에는 각각의 가중치가 존재하는데, 이때 가중치의 값이 크면 클수록 해당 입력 값이 중요하다는 것을 의미합니다.

## 2. 단층 퍼셉트론(Single-Layer Perceptron)
---

위에서 배운 퍼셉트론을 **단층 퍼셉트론**이라고 합니다. 퍼셉트론은 **단층 퍼셉트론**과 **다층 퍼셉트론**으로 나누어지는데, 단층 퍼셉트론은 값을 보내는 단계과 값을 받아서 출력하는 두 단계로만 이루어집니다. 이때 이 각 단계를 보통 층(layer)이라고 부르며, 이 두 개의 층을 **입력층**(input layer)과 **출력층**(output layer)이라고 합니다.

## 3. 다층 퍼셉트론 (MultiLayer Perceptron, MLP)
---

단층 퍼셉트론에서 중간 은닉층을 더 추가하면 **다층 퍼셉트론**(MultiLayer Perceptron, MLP) 라고 합니다. 

이렇게 은닉층이 2개 이상인 신경망을 **심층 신경망**(Deep Neural Network, DNN) 이라고 합니다. 심층 신경망은 다층 퍼셉트론만 이야기 하는 것이 아니라, 여러 변형된 다양한 신경망들도 은닉층이 2개 이상이 되면 심층 신경망이라고 합니다. 
 
입력층과 가까운 층을 보통 **하위 층** 이라 부르고 출력에 가까운 층을 **상위 층** 이라고 부릅니다.

## 4. 전결합층(Fully-connected layer, FC, Dense layer)
---

출력층을 제외하고 모든 층은 학습파라미터를 포함하고 다음층과 완전연결 되어 있습니다. 

다층 퍼셉트론은 은닉층과 출력층에 있는 모든 뉴런은 바로 이전 층의 모든 뉴런과 연결돼 있었습니다. 

이렇게 어떤 층의 모든 뉴런이 이전 층의 모든 뉴런과 연결돼 있는 층을 **전결합층**(Fully-connected layer) 또는 **완전연결층**이라고 합니다. 줄여서 **FC**라고 부르기도 합니다. 동일한 의미로 **밀집층**(Dense layer) 이라고 부르기도 하는데, 케라스에서는 밀집층을 구현할 때 `Dense()`를 사용합니다.

## 6. 활성화 함수 (Activation Function)
---


은닉층과 출력층의 뉴런에서 출력값을 결정하는 함수를 **활성화 함수**(Activation function)라고 합니다.


**선형함수와 비선형함수**

>활성화 함수의 특징은 선형 함수가 아닌 비선형 함수여야 한다는 점입니다. 선형 함수란 출력이 입력의 상수배만큼 변하는 함수를 선형함수라고 합니다. 

인공 신경망에서 활성화 함수는 비선형 함수여야 합니다.만약 선형함수로 층을 계속 쌓는다고 가정해보겠습니다. 첫번째 층은 $y_1=ax_1$, 두번째 층은 $y_2=bx_2$ 라고 합시다. 최종 결과는 $y2=b(ax_1)=bax_1$이 됩니다. $a$와 $b$는 상수기 때문에 $y=Cx$라는 선형함수와 다를게 없습니다.

이와 같은 이유로 대부분의 인공 신경망에선 선형함수와 활성화 함수를 함께 사용해서 한 층을 이룹니다.
하지만 종종 활성화 함수를 사용하지 않는 층을 쓰기도 하는데 학습 가능한 가중치가 새로 생긴다는 점에서 의미가 있습니다.



```python
import numpy as np
import matplotlib.pyplot as plt
```

### (1)  계단 함수(Step function)

<h3>
<center>
$f(x) = 1  \quad if \quad x >0 $
</h3>
</center>


<h3>
<center>
$f(x) = 0  \quad if \quad x <0 $
</h3>
</center>




```python
def step(x):
  return np.array(x>0,dtype=np.int)

x = np.arange(-3,3,0.01)
plt.title("Step fuction")
plt.plot(x, step(x))
plt.show()
```

### (2) 시그모이드 함수(Sigmoid function)

<h3>
<center>
$f(x) = \frac{1}{1+e^{(-x)}}$
</h3>
</center>



```python
def sigmoid(x):
  return 1/(1+np.exp(-x))

x = np.arange(-5,5,0.01)
plt.title("Sigmoid fuction")
plt.plot(x, sigmoid(x))
plt.show()         ##아래와다른점 0,1사이의 값이
```

### (3) 하이퍼볼릭탄젠트 함수(Hyperbolic tangent function)

<h3>
<center>
$f(x) = tanhx = \frac{sinhx}{coshx} = \frac{e^x - e^{-x}}{e^x + e^{-x}}$
</h3>
</center>




```python
x = np.arange(-5,5,0.01)
plt.title("Hyperbolic tangent function")
plt.plot(x, np.tanh(x))
plt.show()                 ###위와 다른점 -1 ,1사이의 값이다
```

### (4) 렐루 함수(ReLU)




인공 신경망의 은닉층에서 가장 인기있는 함수입니다. 0이상의 입력값의 경우에는 미분값이 항상 1입니다. 렐루 함수는 시그모이드 함수와 하이퍼볼릭탄젠트 함수와 같이 어떤 연산이 필요한 것이 아니라 단순 임계값이므로 연산 속도도 빠릅니다.

하지만 여전히 문제점이 존재하는데, 입력값이 음수면 기울기. 즉, 미분값도 0이 됩니다. 그리고 이 뉴런은 다시 회생하는 것이 매우 어렵습니다. 이 문제를 죽은 렐루(dying ReLU)라고 합니다.



<h3>
<center>
$f(x) = max(0,x)$
</h3>
</center>




```python
def relu(x):
  return np.maximum(0,x)
x = np.arange(-5,5,0.01)
plt.title("ReLU")
plt.plot(x, relu(x))
plt.show()                 #연산결과가 0보다 작으면 0 //연산결과가 0보다 클때만 값을 넘긴다
```

### (5) 리키 렐루(Leaky ReLU)




죽은 렐루를 보완하기 위해 ReLU의 변형 함수들이 등장하기 시작했습니다. 변형 함수는 여러 개가 있지만 여기서는 Leaky ReLU에 대해서만 소개합니다. Leaky ReLU는 입력값이 음수일 경우에 0이 아니라 0.001과 같은 매우 작은 수를 반환하도록 되어있습니다. $a$는 하이퍼파라미터로 Leaky('새는') 정도를 결정하며 일반적으로는 0.01의 값을 가집니다. 여기서 말하는 '새는 정도'라는 것은 입력값의 음수일 때의 기울기를 비유하고 있습니다.

<h3>
<center>
$f(x) = max(ax,x)$




```python
def leaky_relu(x,a=0.1):
  return np.maximum(a*x,x)
x = np.arange(-5,5,0.01)
plt.title("leaky_ReLU")
plt.plot(x, leaky_relu(x)) #0이하일때 아주 작은값만 반환
plt.show()   
```

### (6) 소프트맥스 함수(Softamx function)


소프트맥스 함수는 세 가지 이상의 (상호 배타적인) 선택지 중 하나를 고르는 다중 클래스 분류(MultiClass Classification) 문제에 주로 사용됩니다. 

<center>
<h3>
$f(x) = \frac{1}{\sum^d_{j=1}e^{(x_j)}}[e^{x_1}, e^{x_2}, ... e^{x_d}]$



```python
def softmax(x):
  return np.exp(x)/np.sum(np.exp(x))
#x = np.arange(-5,5,0.01)
x=np.array([1,2,3])
plt.title("Softmax function")
plt.plot(x, softmax(x),marker='o')
plt.show()   
```
